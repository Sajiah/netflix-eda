# Cleaning the data

```python
netflix_data.isnull().sum()
```

```python
netflix_data = netflix_data.drop(['duration', 'show_id'], axis=1)
```

```python
netflix_data.head()
```

```python
values = {"director": "NA", "cast": "NA", "country": "NA", "rating": "NA"}
```

Let's handle the missing data

```python
netflix_data = netflix_data.fillna(value=values)
```

```python
netflix_data = netflix_data.fillna({"date_added": netflix_data['date_added'].mode()[0]})
```

```python
netflix_data['date_added']= pd.to_datetime(netflix_data['date_added'])
```

```python
netflix_data.head()
```

```python
netflix_data['year_added'] = pd.DatetimeIndex(netflix_data['date_added']).year
```

```python
df_movie = netflix_df.query('type == "Movie"')
df_movie
```

```python
all_directors = df_movie['director'].str.split(',',expand = True).stack()
all_directors = all_directors.to_frame()
```

```python
all_directors.columns = ['Director']
```

```python
all_directors_total = all_directors.groupby('Director').size().reset_index(name = 'Total Content')
```

```python
all_directors_total
```

```python
all_directors_total = all_directors_total.query("Director != 'NA'").sort_values(by = ['Total Content'] , ascending = False)
```

```python
all_directors_total
```

```python
top_5_movie_directors = all_directors_total.head(5)
```

```python
sn.barplot(x = 'Director', y = 'Total Content', data = top_5_movie_directors).set(title = 'Top 5 Movie Directors')
```

```python
df_show = netflix_df.query('type == "TV Show"')
```

```python
all_directors_show = df_show['director'].str.split(',',expand = True).stack()
all_directors_show = all_directors_show.to_frame()
all_directors_show.columns = ['Director']
```

```python
all_directors_show
```

```python
all_director_total_content = all_directors_show.query("Director != 'NA'").groupby('Director').size().reset_index(name = 'Total Shows')
```

```python
all_director_total_content = all_director_total_content.sort_values(by = 'Total Shows', ascending = False)
```

```python
top_5_show_directors = all_director_total_content.head(5)
```

```python
sn.barplot(x = 'Director', y = 'Total Shows', data = top_5_show_directors).set(title = 'Top 5 show directors')
```

```python
all_listed_mratings = df_movie['listed_in'].str.split(',',expand = True).stack()
all_listed_mratings = all_listed_mratings.to_frame()
all_listed_mratings
```

```python
all_listed_mratings.columns = ['Rating']
```

```python
all_listed_mratings_total = all_listed_mratings.groupby('Rating').size().reset_index(name= 'Total')
all_listed_mratings_total = all_listed_mratings_total.sort_values(by = 'Total', ascending = False)
```

```python
ratings_array = all_listed_mratings['Rating'].unique().tolist()
ratings_array
```

```python
top_5_movie_ratings = all_listed_mratings_total.head(5)
```

```python
# Top 5 movie ratings

sn.barplot(x = 'Rating', y = 'Total', data = top_5_movie_ratings).set(title = 'Top 5 movie ratings')
```

```python
def check_top_5_genre_directors(genre):
    dataset_rating = df_movie[df_movie['listed_in'].str.contains(genre)]
    directors = dataset_rating['director'].str.split(',',expand = True).stack()
    directors = directors.to_frame()
    directors.columns = ['Director']
    total = directors.groupby('Director').size().reset_index(name = 'Total Content')
    top_5_genre = total.query("Director != 'NA'").sort_values(by = ['Total Content'] , ascending = False)
    top_5_genre = top_5_genre.head(5)
    title = '\nTop 5 ' + genre + ' Movie Directors\n'
    plt.hlines(y=top_5_genre['Director'], xmin=0, xmax=top_5_genre['Total Content'], color= '#06837f', alpha=0.7, linewidth=2)
    for i, value in enumerate(top_5_genre['Total Content']):
        plt.text(value, i, value, ha='left', va='center')
    plt.title(title, fontsize=15, weight=600, color='#333d29')
    plt.show()
    sn.despine()
    plt.tight_layout()
    
def check_top_5_genre_directors_show(genre):
    dataset_rating = df_show[df_show['listed_in'].str.contains(genre)]
    directors = dataset_rating['director'].str.split(',',expand = True).stack()
    directors = directors.to_frame()
    directors.columns = ['Director']
    total = directors.groupby('Director').size().reset_index(name = 'Total Content')
    top_5_genre = total.query("Director != 'NA'").sort_values(by = ['Total Content'] , ascending = False)
    top_5_genre = top_5_genre.head(5)
    if top_5_genre.empty:
        print('Empty data for', genre)
    else:
        title = '\nTop 5 ' + genre + ' Show Directors\n'
        plt.hlines(y=top_5_genre['Director'], xmin=0, xmax=top_5_genre['Total Content'], color= '#06837f', alpha=0.7, linewidth=2)
        for i, value in enumerate(top_5_genre['Total Content']):
            plt.text(value, i, value, ha='left', va='center')
        plt.title(title, fontsize=15, weight=600, color='#333d29')
        plt.show()
        sn.despine()
        plt.tight_layout()
    
    
```

```python
for genre in ratings_array:
    check_top_5_genre_directors(genre)
   
```

```python
for genre in ratings_array:
    check_top_5_genre_directors_show(genre)
```
